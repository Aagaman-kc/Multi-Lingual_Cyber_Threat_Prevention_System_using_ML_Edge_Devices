<h2>✅ What DPI Can Detect in the Pipeline</h2>

<table border="1" style="border-collapse: collapse; width:100%;">
    <tr>
        <th>Attack Type</th>
        <th>How DPI Detects It</th>
    </tr>
    <tr>
        <td><b>XSS, SQL Injection, Command Injection</b></td>
        <td>
            Inspect HTTP GET/POST payloads, headers, and URI parameters for suspicious signatures, 
            regex patterns, or anomalous characters and structures.
        </td>
    </tr>
    <tr>
        <td><b>DoS / Volumetric Attacks</b></td>
        <td>
            Detect sudden traffic spikes, abnormal packet rates, and flooding behaviour 
            using flow statistics and traffic monitoring.
        </td>
    </tr>
    <tr>
        <td><b>Unknown / Zero-Day Attacks</b></td>
        <td>
            Use unsupervised anomaly detection techniques such as 
            <code>Isolation Forest</code> to identify traffic patterns 
            that deviate from normal network behaviour.
        </td>
    </tr>
    <tr>
        <td><b>DNS Tunneling &amp; Beaconing</b></td>
        <td>
            Analyze flow-based features such as frequent small requests, 
            periodic communication patterns, and unusual domain entropy.
        </td>
    </tr>
</table>

<h2>❌ Limitations of DPI-Based Detection</h2>

<table border="1" style="border-collapse: collapse; width:100%;">
    <tr>
        <th>Limitation</th>
        <th>Description</th>
    </tr>
    <tr>
        <td><b>Encrypted Traffic (HTTPS / TLS)</b></td>
        <td>
            DPI cannot inspect encrypted payload contents. 
            Only metadata such as packet size, timing, IP addresses, and traffic flow can be analyzed. 
            In such cases, flow-based anomaly detection techniques must be used.
        </td>
    </tr>
    <tr>
        <td><b>Application-Layer Evasion</b></td>
        <td>
            Attackers may evade signature-based detection by splitting malicious payloads across packets, 
            encoding characters, or obfuscating request patterns.
        </td>
    </tr>
    <tr>
        <td><b>High False Positives</b></td>
        <td>
            Network anomaly detection models may incorrectly flag legitimate but unusual traffic 
            such as software updates, backup transfers, or authorized SSH activity as malicious.
        </td>
    </tr>

</table>
