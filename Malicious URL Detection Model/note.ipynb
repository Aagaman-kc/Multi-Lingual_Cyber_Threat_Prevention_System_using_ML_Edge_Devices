<h3>How to Use on Raspberry Pi 5</h3>

<p>Copy these files to your Raspberry Pi:</p>

<ul>
  <li>rf_url_model.pkl</li>
  <li>label_encoder.pkl</li>
  <li>feature_columns.pkl</li>
  <li>predict_url.py</li>
</ul>

<p>Install required packages:</p>

<pre><code>pip install pandas scikit-learn joblib</code></pre>

<p>Run prediction:</p>

<pre><code>python predict_url.py https://example.com/login</code></pre>

<h3>Integration Ideas</h3>

<ul>
  <li>Flask web API for HTTP requests</li>
  <li>Browser extension integration</li>
  <li>Batch URL processing from CSV/log files</li>
  <li>SIEM or firewall integration</li>
</ul>

<h3>Performance Notes</h3>

<p>
The Random Forest model processes a single URL in approximately
10–30 ms on Raspberry Pi 5.
</p>

<p>If higher throughput is needed:</p>

<ul>
  <li>Reduce <code>n_estimators</code> to 50</li>
  <li>Use ONNX Runtime for faster inference</li>
</ul>
